```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import plotly.express as px
warnings.filterwarnings('ignore')
```

## Temp MÃ¡x

```python
df = pd.read_excel('graps.xlsx')
```

```python
df = df.set_index('Date')
```

```python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
fig, ax = plt.subplots(5, 1, sharex=True, sharey=True, figsize=(10, 20))
#ax[4].tick_params(axis='both', which='major', labelsize=11)
#ax[4].tick_params(axis='both', which='minor', labelsize=10)
fig.text(0.06, 0.5, 'Temperatura (Â°C)', ha='center',
         va='center', rotation='vertical', fontsize=15)
#fig.text(0.5, 0.11, 'Data', ha='center', va='center',fontsize = 12)
ax[0].set_title('LDAPS')
sns.lineplot(ax=ax[0], data=df[['PrevisÃ£o LDAPS', 'Temperatura Real']])
ax[1].set_title('SVR')
sns.lineplot(ax=ax[1], data=df[['PrevisÃ£o SVR', 'Temperatura Real']])
ax[2].set_title('RF')
sns.lineplot(ax=ax[2], data=df[['PrevisÃ£o RF', 'Temperatura Real']])
ax[3].set_title('MLP')
sns.lineplot(ax=ax[3], data=df[['PrevisÃ£o MLP', 'Temperatura Real']])
ax[4].set_title('MME')
ax[4].set_xlabel('Data', fontsize=15)
sns.lineplot(ax=ax[4], data=df[['PrevisÃ£o MME', 'Temperatura Real']])
```
